```python
def ecb_reshape(data,metric_name):
    df = data.copy()
    df.drop(data.columns[1],axis=1,inplace=True)
    # Rename columns for simplicity
    df.columns = ['DATE', metric_name]
    # Convert "DATE" to Year-Month format
    #df["DATE"] = pd.to_datetime(df["DATE"]).dt.to_period("M")
    return df
```

```python
m1_re = ecb_reshape(m1,'M1')
m2_re = ecb_reshape(m2,'M2')
m3_re = ecb_reshape(m3,'M3')

df_merged = m1_re.merge(m2_re, on="DATE", how="outer").merge(m3_re, on="DATE", how="outer")
df_merged["DATE"] = pd.to_datetime(df_merged["DATE"]).dt.to_period("M")


df_merged["GEO"] = "EU"
df_merged
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(20, 5))

df_merged.plot(figsize=(12, 6))

plt.xticks(rotation=45)  # Rotate labels 45 degrees
plt.tight_layout()       # Adjust layout to prevent cutoff
plt.show()
```
